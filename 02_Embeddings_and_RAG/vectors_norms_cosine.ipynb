```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib.patches import Arc

# Define two 2D vectors
v1 = np.array([3, 4])
v2 = np.array([4, 1])

# Magnitudes
mag_v1 = np.linalg.norm(v1)
mag_v2 = np.linalg.norm(v2)

# Unit vectors
u1 = v1 / mag_v1
u2 = v2 / mag_v2

# Cosine similarity and angle
cos_sim = np.dot(v1, v2) / (mag_v1 * mag_v2)
angle_rad = math.acos(cos_sim)
angle_deg = np.degrees(angle_rad)

# Plot
plt.figure(figsize=(6,6))

# Original vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1,
           color='b', alpha=0.5, label=f"v1 |v1|={mag_v1:.2f}")
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1,
           color='r', alpha=0.5, label=f"v2 |v2|={mag_v2:.2f}")

# Unit vectors
plt.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1,
           color='b', width=0.005, label="unit v1")
plt.quiver(0, 0, u2[0], u2[1], angles='xy', scale_units='xy', scale=1,
           color='r', width=0.005, label="unit v2")

# Draw angle arc
arc_radius = 1.2   # radius of the arc
start_angle = np.degrees(np.arctan2(u1[1], u1[0]))
end_angle   = np.degrees(np.arctan2(u2[1], u2[0]))
arc = Arc((0,0), arc_radius*2, arc_radius*2, angle=0,
          theta1=min(start_angle, end_angle),
          theta2=max(start_angle, end_angle),
          color="green", linewidth=2)
plt.gca().add_patch(arc)

# Label angle
plt.text(arc_radius*0.8, 0.2, f"θ ≈ {angle_deg:.1f}°",
         fontsize=10, color="green", weight="bold")

# Axes, grid, and style
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.gca().set_aspect('equal')

plt.legend()
plt.title("Cosine Similarity and Angle Between Vectors")
plt.show()
```
